# 3. Building Predictive Models

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,1:]
y = df['Survived']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
x
```

```python
y
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier()
xgb.fit(xtrain,ytrain)
ypred = xgb.predict(xtest)
```

```python
print('Training Accuracy: ', xgb.score(xtrain,ytrain))
print('Testing Accuracy: ', xgb.score(xtest,ytest))
```

# 4. Hyper Parameter Tuning Model

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_estimator' : [90,110,120,150],
    'learning_rate': [0.5,0.6,0.7,0.8,0.9,1],
    'max_depth': [3,4,5,6,7,8,9,11],
    'lambda': [1,2,3]
}
```

# 5. Comparative Analysis:

**Performance Comparison:**

- **LightGBM:** Training Accuracy = 98.39%, Testing Accuracy = 82.02%

- **XGBoost:** Training Accuracy = 95.02%, Testing Accuracy = 79.03%

*- LightGBM has a higher training accuracy but suffers from a larger gap between training and testing accuracy, indicating possible overfitting. XGBoost has a lower training accuracy but generalizes better with a smaller gap*.

**Strengths and Weaknesses:**

- LightGBM is faster and efficient for large datasets but tends to overfit more.

- XGBoost is more robust against overfitting but may be slower in training.

```python
models = ['LightGBM', 'XGBoost']
train_acc = [0.9839, 0.9502]
test_acc = [0.8202, 0.7903]

x = np.arange(len(models))
width = 0.3

fig, ax = plt.subplots()
ax.bar(x - width/2, train_acc, width, label='Training Accuracy')
ax.bar(x + width/2, test_acc, width, label='Testing Accuracy')

ax.set_ylabel('Accuracy')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.show()
```
