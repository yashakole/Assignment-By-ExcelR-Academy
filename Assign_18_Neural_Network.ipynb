```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense # Dense is used to add layers(Hidden, Output)
from tensorflow.keras import Sequential # Used to all layers in sequence to initialize ANN model that is initially random weight will be assigned
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
```

# 1. Data Exploration and Preprocessing

```python
df = pd.read_csv('/content/Alphabets_data.csv')
```

```python
df # Letter is target column
```

```python
df['letter'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df['letter'] = LE.fit_transform(df['letter'])
```

```python
df
```

```python
x = df.drop(columns=['letter'])
y = df['letter']
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
x
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20,random_state=1)
```

# 2. Model Implementation

```python
model = Sequential() # Sequential is class in Keras that allows you to build ANN model layer by layer and created Instance # Input layer
```

```python
model.add(Dense(units=16, activation = 'relu')) # Added Hidden Layer
```

```python
model.add(Dense(units=1,activation = 'sigmoid')) # Added Output layer
```

```python
model.compile(optimizer='adadelta', loss = 'binary_crossentropy', metrics = ['accuracy'])# Establish the connection between layers
```

```python
model.fit(xtrain,ytrain, epochs=100) # fit the data, perform forward and backward propagation
```

```python
ypred = model.predict(xtest)
```

```python
ypred = ypred>0.5
```

```python
print(classification_report(ytest,ypred))
```

```python
import keras_tuner as kt
```

```python
def hyper(hp):
    model = Sequential()
    # Dynamically set input dimension
    model.add(Dense(
        units=hp.Int('unit0', min_value=34, max_value=256, step=32),
        activation=hp.Choice('activation0', values=['relu', 'tanh', 'sigmoid']),
        input_dim=xtrain.shape[1]  # Use actual feature count from xtrain
    ))
    # Add hidden layers
    for i in range(1, hp.Int('num_layers', min_value=1, max_value=10)):
        model.add(Dense(
            units=hp.Int('units' + str(i), min_value=32, max_value=256, step=32),
            activation=hp.Choice('activation' + str(i), values=['relu', 'tanh', 'sigmoid'])
        ))
    # Output layer
    model.add(Dense(units=1, activation='softmax'))

    # Compile
    model.compile(
        optimizer=hp.Choice('optimizer', values=['adam', 'rmsprop', 'sgd']),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
tuner = kt.RandomSearch(
    hyper,
    objective='val_accuracy',
    max_trials=3,
    directory='my_dir',  # Optional: Save results
    project_name='kt_search'  # Optional: Project name
)
```

```python
tuner.search(xtrain, ytrain, epochs=3, validation_data=(xtest, ytest))
```

```python
tuned_model = tuner.get_best_models(num_models=1)[0]
```

```python
tuned_model.fit(xtrain,ytrain,epochs=50,validation_data=(xtest,ytest))
```

```python
tuned_model.evaluate(xtrain,ytrain)
```
