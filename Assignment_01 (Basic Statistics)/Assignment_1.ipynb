```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

Data Visualization

```python
num_df = df._get_numeric_data()
print(num_df.columns)
```

```python
for i in num_df:
  plt.figure()
  df.hist(i)
```

```python
for i in num_df:
    plt.figure()
    plt.boxplot(df[i])
    plt.title(i)
```

```python
df.info()
```

```python
obj_col = df.select_dtypes(include=['object']).columns
obj_col
```

```python
for i in obj_col:
  plt.figure(figsize=(8,6))
  plt.title(i)
  plt.bar(df.index,df[i])
```

Standardization/ Z-Score

```python
scaler = StandardScaler()
scaler_df = scaler.fit_transform(num_df)
scaler_df = pd.DataFrame(scaler_df, columns=num_df.columns)
scaler_df
```

Get Dummies

```python
df_encoded = pd.get_dummies(df, columns=obj_col.tolist(), dtype=int)
df_encoded
```
